```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

movie = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
movies = pd.merge(ratings, movie, on='movieId')
```

```python
movies = movies[['movieId', 'title', 'userId', 'rating']]
movies
```

```python
table = pd.pivot_table(movies, values='rating', index='userId', columns='title')
```

```python
table
```

```python
table.fillna(0, inplace=True)
```

```python
table
```

```python
userMeans = table.mean(axis=1)
```

```python
userMeans
```

```python
meanCenteredPrediction = table.subtract(userMeans, axis=0)
```

```python
meanCenteredPrediction
```

```python
userSimilarity = cosine_similarity(table)
```

```python
userSimilarity
```

```python
def MeanPredictionFuntion(user_id, movieName):
    movieRatings = table[movieName]
    similarityScores = userSimilarity[user_id-1]
    dotProduct = np.dot(similarityScores, movieRatings)
    sumOfScores = np.sum(similarityScores)
    prediction = dotProduct / abs(sumOfScores)
    return prediction
```
