```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
```

```python
movie
```

```python
ratings
```

```python
movies = pd.merge(movie, ratings, on='movieId')
```

```python
movies
```

```python
moviesMean = pd.DataFrame(movies.groupby('movieId')['rating'].mean())
```

```python
moviesMean
```

```python
movies = pd.merge(movies, moviesMean, on='movieId')
```

```python
movies
```

```python
genres = movies['genres'].str.get_dummies('|')
features = pd.concat([movies[['movieId', 'rating_x']], genres], axis=1)
features.set_index('movieId', inplace=True)
```

```python
genres
```

```python
features
```

```python
similarities = cosine_similarity(features)
```

```python
similarities
```

```python
def similar3Movies(movieName, n=3):
    movieId = movies[movies['title'].str.contains(movieName)].index[0]
    scores = list(enumerate(similarities[movieId]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    scores = scores[1:n+1]
    movieIndex = [i[0] for i in scores]
    return movies['title'].iloc[movieIndex]
```

```python
movieInput = input("Enter Movie Name: ")
```

```python
print(similar3Movies(movieInput))
```
